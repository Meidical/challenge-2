```python
import ahpy

# Comparações par-a-par (escala de Saaty)
# Intensity of importance | Description
# ------------------------|---------------------------------------
# 9                       | Extreme importance
# 7                       | Very strong or demonstrated importance
# 5                       | Strong importance
# 3                       | Moderate importance
# 1                       | Equal importance
# 2,4,6,8                 | Intermediate values



# Regra clássica: CR < 0.10 → aceitável
# Regra exigente (saúde): CR < 0.05 → muito bom
# CR ≈ 0.01 → quase perfeito



# 1. Comparações entre CRITÉRIOS (nível 2)
comparisons_criteria = {
    ('Market', 'Income'): 1/5,
    ('Market', 'Infrastructure'): 3,
    ('Market', 'Transportation'): 4,
    ('Income', 'Infrastructure'): 9,
    ('Income', 'Transportation'): 7,
    ('Infrastructure', 'Transportation'): 2
}

criteria = ahpy.Compare(
    name='Criteria',
    comparisons=comparisons_criteria,
    precision=4,
    random_index='saaty'
)

print("Pesos dos critérios:")
print(criteria.target_weights)

print("\nCR (critérios):")
print(criteria.consistency_ratio)


# 2. Comparações das ALTERNATIVAS por critério (nível 3)
comparisons_market = {
    ('Porto', 'Braganca'): 3,
    ('Porto', 'Lisboa'): 2,
    ('Lisboa', 'Braganca'): 5
}

comparisons_income = {
    ('Porto', 'Braganca'): 6,
    ('Lisboa', 'Porto'): 3,
    ('Lisboa', 'Braganca'): 9
}

comparisons_infrastructure = {
    ('Braganca', 'Porto'): 3,
    ('Braganca', 'Lisboa'): 7,
    ('Porto', 'Lisboa'): 1
}

comparisons_transportation = {
    ('Braganca', 'Porto'): 3,
    ('Braganca', 'Lisboa'): 4,
    ('Porto', 'Lisboa'): 1/2
}

alt_market = ahpy.Compare('Market', comparisons_market, precision=4, random_index='saaty')
alt_income = ahpy.Compare('Income', comparisons_income, precision=4, random_index='saaty')
alt_infra = ahpy.Compare('Infrastructure', comparisons_infrastructure, precision=4, random_index='saaty')
alt_trans = ahpy.Compare('Transportation', comparisons_transportation, precision=4, random_index='saaty')

print("\nCR (alternativas por critério):")
print("Market:", alt_market.consistency_ratio)
print("Income:", alt_income.consistency_ratio)
print("Infrastructure:", alt_infra.consistency_ratio)
print("Transportation:", alt_trans.consistency_ratio)


# 3. Ligar a hierarquia e calcular o ranking final
criteria.add_children([alt_market, alt_income, alt_infra, alt_trans])

print("\nRanking final das alternativas:")
print(criteria.report(show=True))
```
