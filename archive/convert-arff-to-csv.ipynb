## Imports

```python
from scipy.io import arff
import pandas as pd

ARFF_PATH = 'raw-datasets/raw-bone-marrow-original.arff'
CSV_PATH = 'raw-datasets/raw-bone-marrow.csv'
```

## Reading

```python
with open(ARFF_PATH, encoding="utf-8") as arff_file:
    original_dset = arff.loadarff(arff_file)
```

## Conversion

```python
df = pd.DataFrame(original_dset[0])
df
```

## Basic Cleaning

```python
df.info()
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.decode('utf-8')
```

```python
df
```

## Writing

```python
df.to_csv(CSV_PATH, index=False)
```
